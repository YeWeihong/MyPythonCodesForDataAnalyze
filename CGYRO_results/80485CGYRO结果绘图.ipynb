# cs/a 计算

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
计算等离子体声速 cs 和 cs/a
适用于 EAST 装置 (a = 0.45 m)，氘等离子体
公式：cs = sqrt( γ (Te + Ti) / mi )
"""

import math

# 物理常数
gamma = 5.0 / 3.0              # 绝热指数（常用值）
a_minor = 0.45                 # EAST 小半径 (m)

# 单位转换常数
keV_to_J = 1.60217662e-16      # 1 keV = 1.602e-16 J
u = 1.66053906660e-27          # 原子质量单位 (kg)
mi_deuterium = 2.0 * u         # 氘离子质量 ≈ 2 u (标准近似)

# 更方便的实用公式常数（cs in m/s, T in keV, A=2 for deuterium）
CS_FACTOR = 9.79e5             # ≈ sqrt( keV_to_J / u ) * sqrt(3/2 * 1e3) 近似值

def calculate_cs(Te_keV, Ti_keV=None, adiabatic_index=5/3, A=2.0):
    """
    计算等离子体声速 cs (m/s)
    参数:
        Te_keV: 电子温度 (keV)
        Ti_keV: 离子温度 (keV)，若为None则假设 Ti = Te
        adiabatic_index: γ，默认 5/3
        A: 离子质量数，氘=2，氢=1，氚=3
    """
    if Ti_keV is None:
        Ti_keV = Te_keV
    
    Te_plus_Ti = Te_keV + Ti_keV
    
    # 方法1：实用公式（最常用，精度足够）
    cs = CS_FACTOR * math.sqrt(Te_plus_Ti / A)
    
    # 方法2：显式计算（用于验证）
    # pressure_term = adiabatic_index * Te_plus_Ti * keV_to_J
    # cs2 = pressure_term / (mi_deuterium * A / 2.0)   # A=2 时 mi = 2u
    # cs_explicit = math.sqrt(cs2)
    # print(f"  显式计算 cs = {cs_explicit:.3e} m/s (用于对比)")
    
    return cs


def main():
    print("=== EAST 等离子体 cs / a 计算工具 ===\n")
    print(f"装置小半径 a = {a_minor} m")
    print(f"离子种类：氘 (A=2)\n")
    
    # 获取输入
    temp_input = input("请输入温度值 (keV)：\n"
                      "  - 输入一个值 → 假设 Te = Ti = 该值\n"
                      "  - 输入两个值（用空格分隔） → Te Ti\n"
                      "你的输入： ").strip()
    
    if ' ' in temp_input:
        try:
            Te, Ti = map(float, temp_input.split())
        except:
            print("输入格式错误，使用默认值 Te=Ti=0.2 keV")
            Te = Ti = 0.2
    else:
        try:
            Te = Ti = float(temp_input)
        except:
            print("输入错误，使用默认值 0.2 keV")
            Te = Ti = 0.2
    
    cs = calculate_cs(Te, Ti)
    
    cs_over_a = cs / a_minor
    tau_sound_us = (a_minor / cs) * 1e6   # 转换为微秒
    
    print("\n" + "="*50)
    print(f"Te          = {Te:6.3f} keV")
    print(f"Ti          = {Ti:6.3f} keV")
    print(f"Te + Ti     = {Te+Ti:6.3f} keV")
    print(f"声速 cs     = {cs:.3e} m/s")
    print(f"cs / a      = {cs_over_a:.3e} s⁻¹")
    print(f"a / cs      = {tau_sound_us:.3f} μs  (声学穿越时间)")
    print("="*50)


if __name__ == "__main__":
    main()
```
